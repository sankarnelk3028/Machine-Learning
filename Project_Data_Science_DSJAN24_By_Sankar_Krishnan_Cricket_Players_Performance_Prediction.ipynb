```python
#replace object values in all columns with NaN and change all dtypes to float
df = df.replace('-', np.nan)
df = df.replace('DNB',np.nan)
df = df.replace('TDNB',np.nan)
df = df.astype(float)
df.dtypes
```

```python
df= pd.concat([df,dfr["Performance"]],axis=1)
df
```

```python
x=df.drop('Performance',axis=1)
x
```

```python
y=df['Performance']
y
```

```python
# Train_Test_Split
from sklearn.model_selection import train_test_split,GridSearchCV
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30,random_state=42)
x_train
```

```python
# Normalization

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
# Importing Classifiers:
# 1. K-Nearest Neighbors classifier,2. Gaussian Naive Bayes classifier
# 3. Support Vector Classifier,4.  Decision Tree classifier,5. Random Forest classifier
#
# Importing Metrics
# confusion_matrix, ConfusionMatrixDisplay: Tools for creating and displaying a confusion matrix.
# accuracy_score: Metric to calculate the accuracy of predictions.
# classification_report: Tool to create a detailed classification report with precision, recall, and F1-score.
#
# Initializing Classifiers
# K-Nearest Neighbors (KNN): Initialized with 7 neighbors.
# Gaussian Naive Bayes (GNB): Initialized with default parameters.
# Support Vector Classifier (SVC): Initialized with default parameters.
# Decision Tree Classifier (DTC): Initialized with 'entropy' as the criterion for splitting nodes
# Random Forest Classifier (RFC): Initialized with default parameters.
#
# Creating a List of Classifiers
# A list lst is created to store all the initialized classifiers for easy iteration and comparison.
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,accuracy_score,classification_report
knn=KNeighborsClassifier(n_neighbors=7)
nav= GaussianNB()
sv=SVC()
dtc=DecisionTreeClassifier(criterion='entropy')
rf=RandomForestClassifier()
lst=[knn,nav,sv,dtc,rf]
```

```python
# This code iterates through each classifier, trains it on the training data, predicts labels for the test data, and evaluates the model's performance using various metrics.
# It prints and displays the results for each model, providing a comprehensive overview of how each classifier performs on the given dataset.
```

```python
print('-----------------------Model Creation-----------------------')
print('')
print('Models Created Here')
print(' ')
print('1. KNN Classifier\n'
      '2. Naive Bayes Classifier\n'
      '3. Support Vector Classifier\n'
      '4. Decision Tree Classifier\n'
      '5. Random Forest Classifier')
print(' ')
for i in lst:
    print("Model",i," started")
    print(' ')
    print(i)
    print(' ')
    i.fit(x_train, y_train)
    print('Predicted Value')
    print(' ')
    y_pred = i.predict(x_test)
    print(' ')
    print(y_pred)
    print(' ')
    print('Confusion matrix is')
    print(' ')
    print(confusion_matrix(y_test, y_pred))
    print(' ')
    print('Accuraccy Score is')
    print(' ')
    print(accuracy_score(y_test, y_pred))
    print(' ')
    print('Classification report is')
    print(' ')
    print(classification_report(y_test, y_pred))
    print(' ')
    cmd = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred))
    cmd.plot()
    plt.show()
    print('')
    print('-----------------------Model Ended-----------------------')
    print(' ')
```
